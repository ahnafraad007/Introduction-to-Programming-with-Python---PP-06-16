```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading The Dataset
data = pd.read_csv('/content/titanic.csv')

# Question 01: Determining the Survival Rate of Passengers Aged Under 5
Age_Under_5 = data[data['Age'] < 5]
Survival_Rate_Under_5 = Age_Under_5['Survived'].mean()
print("{:^80}".format("Answer to the Question no. 01\n"))
print("Survival Rate of Passengers Aged Under 5 is {:.3%}".format(Survival_Rate_Under_5))
print("\n")
print("\n")
print("\n")

# Question 02: Plotting Count Plot of Embarked with Survival as the Hue
print("{:^80}".format("Answer to the Question no. 02\n\n\n"))
Embarked_Counts = data.groupby(['Embarked', 'Survived']).size().unstack(fill_value=0)
X_labels = Embarked_Counts.index
Not_Survived = Embarked_Counts[0]
Survived = Embarked_Counts[1]

# Plotting
X = np.arange(len(X_labels))
width = 0.3
plt.bar(X - width / 2, Not_Survived, width, label="Not Survived", color='red', alpha=1)
plt.bar(X + width / 2, Survived, width, label="Survived", color='blue', alpha=1)
plt.xticks(X, X_labels)
plt.title('Embarkation vs Survival')
plt.xlabel('Port of Embarkation')
plt.ylabel('Passenger Count')
plt.legend()
plt.show()
print("\n")
print("\n")
print("\n")

# Question 03: Calculation of the Average Fare Paid by Passengers Aged Above 50
print("{:^80}".format("Answer to the Question no. 03\n"))
Above_50 = data[data['Age'] > 50]
Average_Fare_Above_50 = Above_50['Fare'].mean()
print("Average Fare Paid by Passengers Aged above 50 is ${:.3f}".format(Average_Fare_Above_50))
print("\n")
print("\n")
print("\n")

# Question 04: Comparison of Survival Rates for Passengers with Different Family Sizes Using Bar Plot
print("{:^80}".format("Answer to the Question no. 04\n\n\n"))
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
Family_Survival_Rate = data.groupby('FamilySize')['Survived'].mean()

# Plotting
plt.bar(Family_Survival_Rate.index, Family_Survival_Rate.values, color='green', alpha=1, edgecolor='black')
plt.title('Survival Rates by Family Size')
plt.xlabel('Family Size(Number of Members)')
plt.ylabel('Survival Rate')
plt.grid(axis='y', linestyle='--', alpha=1)
plt.xticks(Family_Survival_Rate.index)
plt.show()
print("\n")
print("\n")
print("\n")

# Question 05: Identifying the Proportion of Passengers in Each Pclass Who Survived
print("{:^80}".format("Answer to the Question no. 05\n"))
Class_Survival = data.groupby('Pclass')['Survived'].mean()
print("Proportion of passengers Who survived by Class:\n")
for pclass, survival_rate in Class_Survival.items():
    print(f"   Class {pclass}: {survival_rate:.3%}")
    print("\n")
print("\n")
print("\n")

# Question 06: KDE Plot for the Age Distribution of Survivors
print("{:^80}".format("Answer to the Question no. 06\n\n\n"))
Survivors_Age = data[data['Survived'] == 1]['Age'].dropna()
Ages = Survivors_Age.values

# Histogram
plt.hist(Ages, bins=30, density=True, color='lightgreen', alpha=1, edgecolor='black')
plt.title('Age Distribution of The Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=1)
plt.show()
```
